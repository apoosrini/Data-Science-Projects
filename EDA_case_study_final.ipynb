# Data Cleaning and Manipulation

```python
#Finding CODE_GENDER has incorrect value XNA for 4 rows
app.CODE_GENDER.value_counts()
```

```python
#updating XNA with Mode values for CODE_GENDER as its only 4 rows
app.CODE_GENDER= app["CODE_GENDER"].str.replace('/',' ')
app.CODE_GENDER = app.CODE_GENDER.replace('XNA','F')
app.CODE_GENDER.value_counts()
```

```python
#Finding columns having null values percentage above 50% and dropping them as those are not useful for analysis
#Also Printing all columns which are dropped.
for i in app.columns:
    percent=app[i].isna().mean().round(4)*100
    if percent>=50:
        print(i,app[i].isnull().sum(),percent)   
        app.drop(i,axis=1,inplace=True)
```

```python
#Verifying size after dropping columns in above steps
app.shape
```

```python
#Finding columns having null values percentage between 15% to 50%
for i in app.columns:
    percent=app[i].isna().mean().round(4)*100
    if percent>=15 and percent < 50:
        print(i,app[i].isnull().sum(),percent)       
```

```python
# Updating null values with median for EXT_SOURCE_3
app.EXT_SOURCE_3 = app.EXT_SOURCE_3.replace(np.nan,app.EXT_SOURCE_3.median())
app.EXT_SOURCE_3 = round(app.EXT_SOURCE_3 ,2)
app.EXT_SOURCE_3.isnull().sum()
```

```python
#updating null values with Mode values for coulmns OCCUPATION_TYPE
app.OCCUPATION_TYPE = app.OCCUPATION_TYPE.replace(np.nan,"Unknown")
app.OCCUPATION_TYPE.isnull().sum()
app.OCCUPATION_TYPE.value_counts()
```

```python
#Finding columns having null values percentage below 15% 
#No action done for these columns. Will consider to remove if its needed
for i in app.columns:
    tot=(app[i].isnull().sum()/app[i].index.size)*100
    if tot > 0 and tot < 15:
        print(i,app[i].isnull().sum(),tot)
```

```python
#convert DAYS_BIRTH TO AGE in years and drop column DAYS_BIRTH
app['AGE']=round(app['DAYS_BIRTH'] * -0.002738).astype(int) #convert to int
app['AGE'].head()
app = app.drop(['DAYS_BIRTH'],axis=1)
```

```python
#convert DAYS_EMPLOYED TO EMPLOYED_AGE in years and drop column DAYS_EMPLOYED
app['YEARS_EMPLOYED']=round(app['DAYS_EMPLOYED'] * -0.002738).astype(int) #convert to int
app=app.drop(['DAYS_EMPLOYED'],axis=1)
```

```python
#There are 2 unknown values for Family status
app.NAME_FAMILY_STATUS.value_counts()
```

```python
# Replacing Unknown with mode value: Married
app.NAME_FAMILY_STATUS = app.NAME_FAMILY_STATUS.replace('Unknown','Married')
```

## Outlier check & Analysis

```python
# Selected top most columns useful for analysis and stored in list for further analyis.
# storing numeric types into cols_num and categorical into cols_cat
# Plotting box plot and bar plot for each column to notice outliers 
# Code is written using for loop by iterating over each column
from pandas.api.types import is_numeric_dtype

inputList = ['NAME_CONTRACT_TYPE','CODE_GENDER',
'NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE','AMT_INCOME_TOTAL',
 'AMT_CREDIT','OCCUPATION_TYPE','AGE','EXT_SOURCE_2', 'EXT_SOURCE_3','NAME_FAMILY_STATUS',
'NAME_HOUSING_TYPE','FLAG_MOBIL','FLAG_CONT_MOBILE','REG_REGION_NOT_LIVE_REGION']            

cols_num=[]
cols_cat=[]
for i in inputList:
    if is_numeric_dtype(app[i]):
        #print (app[i].dtype)
        cols_num.append(i)
    else:
        cols_cat.append(i)
    
print("cols_num",cols_num)
print("cols_cat", cols_cat)
       
```

# Plot univariate graphs for categorical data

```python
#Plot univariate graphs for categorical variables
a = 4  # number of rows
b = 2  # number of columns
c = 1  # initialize plot counter

fig = plt.figure(figsize=(16,25))

for i in cols_cat:
    plt.subplot(a, b, c)
    plt.title('{}, subplot: {}{}{}'.format(i, a, b, c))
    plt.xlabel(i)
    plt.tight_layout()
    plt.xticks(rotation=90)
    sns.countplot(app[i])
    c = c + 1

plt.show()
```

# Plot univariate graphs for categorical variables

```python
#Plot univariate graphs for Numerical data
a = 8  # number of rows
b = 2  # number of columns
c = 1  # initialize plot counter

fig = plt.figure(figsize=(14,30))

for i in cols_num:
    plt.tight_layout()
    plt.subplot(a, b, c)
    plt.title('{} (dist), subplot: {}{}{}'.format(i, a, b, c))
    plt.xlabel(i)
    sns.distplot(app[i])
    c = c + 1

    plt.subplot(a, b, c)
    plt.title('{} (box), subplot: {}{}{}'.format(i, a, b, c))
    plt.xlabel(i)
    plt.boxplot(x = app[i])
    c = c + 1

plt.show()
```
